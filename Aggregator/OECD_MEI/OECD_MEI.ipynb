```python
import pandas as pd

from dbnomics import fetch_series

import warnings
warnings.filterwarnings("ignore", category=FutureWarning)
```

```python
dataframes = [fetch_series("OECD", "MEI", dimensions={"SUBJECT": ["MABMM301",'MANMM101'], "MEASURE": ["STSA"], "FREQUENCY": ["A"]}, max_nb_series=1000000),
              fetch_series("OECD", "MEI", dimensions={"SUBJECT": ["CCRETT01",'IRLTLT01','IRSTCB01'], "FREQUENCY": ["A"]}, max_nb_series=1000000)]

df = pd.concat(dataframes,ignore_index=True)
df.head(20)
```

```python
df.to_csv('OECD_MEI.csv')
```

```python
df.to_pickle("OECD_MEI.pkl")
```
