```python
import pandas as pd

from dbnomics import fetch_series

import warnings
warnings.filterwarnings("ignore", category=FutureWarning)
```

```python
datasets=['PIBN','PIBC','IMECO','IHPC','BDP4','DPE','SIM','AM_A','TOFE']
```

```python
df = pd.concat([fetch_series("BCEAO", dataset, max_nb_series=10000) for dataset in datasets], ignore_index=True)
df.sort_values(by=['period','country'], inplace=True)
df.head(20)
```

```python
df.to_csv('BCEAO.csv')
```

```python
df.to_pickle("BCEAO.pkl")
```
