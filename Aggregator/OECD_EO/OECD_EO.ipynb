```python
import pandas as pd

from dbnomics import fetch_series

import warnings
warnings.filterwarnings("ignore", category=FutureWarning)
```

```python
variables = ['GDP','GDPV','CPIH','POP','IT','CP','CG','XGS','XGSD','MGS','MGSD','EXCH','EXCHER','CBGDPR','GGFLQ','IRS','IRCB','UNR','CPIH_YTYPCT','ITISK']
indicators = ['TES13S','TRS13S','D2D5D91RS13S','B9S13S']
```

```python
dataframes = [fetch_series("OECD", "EO", dimensions={"VARIABLE": variables, 'FREQUENCY':['A']}, max_nb_series=1000000)
              .drop(columns=["FREQUENCY",'Frequency']).rename(columns={"VARIABLE": "INDICATOR",'Variable':'Indicator'}),
              fetch_series("OECD", "NAAG", dimensions={"INDICATOR": indicators}, max_nb_series=1000000)]
df = pd.concat(dataframes,ignore_index=True)
df.head(20)
```

```python
df.to_csv('OECD_EO.csv')
```
