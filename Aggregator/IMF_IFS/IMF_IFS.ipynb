```python
import pandas as pd

from dbnomics import fetch_series

import warnings
warnings.filterwarnings("ignore", category=FutureWarning)
```

```python
indicators = [
    "NGDP_R_XDC", "NC_R_XDC",  
    "NGDP_XDC", "ENDA_XDC_USD_RATE", "EREER_IX", "FPOLM_PA", "BCAXF_BP6_USD", "LP_PE_NUM",
    "LUR_PT", "BGS_BP6_USD", "NFI_XDC", "NI_XDC", "NM_XDC", "NX_XDC", "NC_XDC",
    "PCPI_IX", "PCPI_PC_CP_A_PT", "EDNE_USD_XDC_RATE"
]

master_df = pd.DataFrame()

for indicator in indicators:
    try:
        df = fetch_series("IMF", "IFS", dimensions={"FREQ": ["A"], "INDICATOR": [indicator]}, max_nb_series=1000000)
        df["indicator"] = indicator  
        master_df = pd.concat([master_df, df], ignore_index=True)
    except Exception:
        print(f"Failed to fetch data for {indicator}")

master_df["ref_area"] = master_df.get("REF_AREA", master_df.get("ref_area", None))
master_df = master_df[["period", "value", "indicator", "ref_area", "series_name"]]
```
