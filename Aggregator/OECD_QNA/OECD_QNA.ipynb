```python
import pandas as pd

from dbnomics import fetch_series

import warnings
warnings.filterwarnings("ignore", category=FutureWarning)
```

```python
subjects = ['P5','P51','P3','B1_GS1']
```

```python
df = fetch_series("OECD", "QNA", dimensions={"SUBJECT": subjects,'FREQUENCY':['A'],'MEASURE':['CQR']}, max_nb_series=1000000)
df = df.reset_index(drop=True)
df.head(20)
```

```python
df.to_excel('OECD_QNA.xlsx')
```

```python
df.to_pickle("OECD_QNA.pkl")
```
