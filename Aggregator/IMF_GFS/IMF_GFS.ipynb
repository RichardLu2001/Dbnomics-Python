```python
dataframes = [fetch_series("IMF", "GFSMAB", dimensions={'CLASSIFICATION': ['GNLB__Z'], 'REF_SECTOR': ['S1311B'], 'UNIT_MEASURE': ['XDC_R_B1GQ']}, max_nb_series=1000000),
    fetch_series("IMF", "GFSMAB", dimensions={'CLASSIFICATION': ['G11__Z'], 'REF_SECTOR': ['S1311B'], 'UNIT_MEASURE': ['XDC']}, max_nb_series=1000000),
    fetch_series("IMF", "GFSMAB", dimensions={'CLASSIFICATION': ['G2M__Z'], 'REF_SECTOR': ['S1311B'], 'UNIT_MEASURE': ['XDC']}, max_nb_series=1000000),
    fetch_series("IMF", "GFSMAB", dimensions={'CLASSIFICATION': ['G1__Z'], 'REF_SECTOR': ['S1311B'], 'UNIT_MEASURE': ['XDC']}, max_nb_series=1000000),
    fetch_series("IMF", "GFSMAB", dimensions={'CLASSIFICATION': ['GNLB__Z'], 'REF_SECTOR': ['S1311'], 'UNIT_MEASURE': ['XDC_R_B1GQ']}, max_nb_series=1000000),
    fetch_series("IMF", "GFSMAB", dimensions={'CLASSIFICATION': ['G11__Z'], 'REF_SECTOR': ['S1311'], 'UNIT_MEASURE': ['XDC']}, max_nb_series=1000000),
    fetch_series("IMF", "GFSMAB", dimensions={'CLASSIFICATION': ['G2M__Z'], 'REF_SECTOR': ['S1311'], 'UNIT_MEASURE': ['XDC']}, max_nb_series=1000000),
    fetch_series("IMF", "GFSMAB", dimensions={'CLASSIFICATION': ['G1__Z'], 'REF_SECTOR': ['S1311'], 'UNIT_MEASURE': ['XDC']}, max_nb_series=1000000)]
```
