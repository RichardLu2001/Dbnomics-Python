```python
import pandas as pd

from dbnomics import fetch_series

import warnings
warnings.filterwarnings("ignore", category=FutureWarning)
```

```python
dataframes= [fetch_series("ECB", "IDCM", dimensions={'FREQ': ['A'], 'STO': ['B1GQ'], 'PRICES': ['V','Q'], 'UNIT_MEASURE': ['XDC']}, max_nb_series=1000000),
    fetch_series("ECB", "IDCM", dimensions={'FREQ': ['A'], 'STO': ['B9'], 'UNIT_MEASURE': ['XDC']}, max_nb_series=1000000),
    fetch_series("ECB", "IDCM", dimensions={'FREQ': ['A'], 'STO': ['EMP','P51C','B8N'], 'REF_SECTOR': ['S1']}, max_nb_series=1000000),
    fetch_series("ECB", "IDCM", dimensions={'FREQ': ['A'], 'STO': ['P3','P5','P51G','P6','P7'], 'REF_SECTOR': ['S1'], 'PRICES': ['V'], 'UNIT_MEASURE': ['XDC']}, max_nb_series=1000000)]
```

```python
df = pd.concat(dataframes,ignore_index=True)
df.head(20)
```

```python
df.to_csv('IDCM.csv')
```
