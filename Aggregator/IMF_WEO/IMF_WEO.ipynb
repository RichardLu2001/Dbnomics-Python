```python
import pandas as pd

from dbnomics import fetch_series

import warnings
warnings.filterwarnings("ignore", category=FutureWarning)
```

```python
version = "2025-04"

all_subjects = [
    "NGDP", "NGDP_R", "NGDPRPC", "NID_NGDP", "PCPI", "TM_RPCH", "TX_RPCH",
    "LUR", "LP", "GGR_NGDP", "GGX_NGDP", "GGXCNL_NGDP", "GGSB_NPGDP",
    "GGXWDG_NGDP", "BCA_NGDPD"
]

def add_subject(subject, df):
    df["weo_subject"] = subject
    return df
    
dataframes = [add_subject(subject, fetch_series("IMF", f"WEO:{version}", dimensions={"weo-subject": [subject]}, max_nb_series=1000000))
             for subject in all_subjects]

df = pd.concat(dataframes, ignore_index=True)

keep_cols = [
    "period", "value", "dataset_code", "dataset_name", "unit", "weo_country",
    "weo_subject", "provider_code", "series_code", "series_name"
]
df = df[[col for col in keep_cols if col in df.columns]]
df.head(20)
```

```python
df.to_csv('IMF_WEO.csv')
```
