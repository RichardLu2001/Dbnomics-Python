```python
import pandas as pd

from dbnomics import fetch_series

import warnings
warnings.filterwarnings("ignore", category=FutureWarning)
```

```python
df = fetch_series("BIS", "WS_SPP", dimensions={"UNIT_MEASURE": ['628'], 'VALUE':['N']}, max_nb_series=1000000)
df = df.reset_index(drop=True)
cols = ['period', 'value', 'dataset_name', 'FREQ', 'REF_AREA', 'series_name', 'series_code']
df = df[cols]
df.head(20)
```

```python
df.to_csv('BIS_HPI.csv')
```
