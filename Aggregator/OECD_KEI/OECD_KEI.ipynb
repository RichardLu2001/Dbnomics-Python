```python
import pandas as pd

from dbnomics import fetch_series

import warnings
warnings.filterwarnings("ignore", category=FutureWarning)
```

```python
subjects = ['CPALTT01','IR3TIB01','IRLTLT01','B6BLTT02','LRHUTTTT']
```

```python
df = fetch_series("OECD", "KEI", dimensions={"SUBJECT": subjects, 'FREQUENCY':['A']}, max_nb_series=1000000)
df = df.reset_index(drop=True)
df.head(20)
```

```python
df.to_excel('OECD_KEI.xlsx')
```
