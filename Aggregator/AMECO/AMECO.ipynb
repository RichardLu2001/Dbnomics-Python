```python
import pandas as pd

from dbnomics import fetch_series

import warnings
warnings.filterwarnings("ignore", category=FutureWarning)
```

```python
datasets=['XUNRQ-1','ISN','ILN','UXGS','UMGS','UCNT','UIGT','UITT','UVGD','OVGD','NPTD','NUTN','ZCPIN']
```

```python
df_all = []

for dataset in datasets:
    if dataset in ['ISN','ILN']:
        dimension = {'unit':['-']}
    else:
        dimension = {}
    df_all.append(fetch_series("AMECO", dataset, dimensions=dimension, max_nb_series=1000000))
    
df = pd.concat(df_all,ignore_index=True)
df.head(20)
```

```python
df.to_csv('AMECO.csv')
```

```python
df.to_pickle("AMECO.pkl")
```
