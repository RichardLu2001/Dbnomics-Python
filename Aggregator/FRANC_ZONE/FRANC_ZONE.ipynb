```python
import pandas as pd

from dbnomics import fetch_series

import warnings
warnings.filterwarnings("ignore", category=FutureWarning)
```

```python
indicators = ['gdp_FCFA','gdp_KMF','investment','money_FCFA','money_KMF','price_index_percent','budget_balance_percent']
```

```python
df = fetch_series("Franc-zone", "FRANCZONE", dimensions={"indicator": indicators}, max_nb_series=1000000)
df = df.reset_index(drop=True)
df.head(20)
```

```python
df.sort_values(['period', 'country'], inplace=True)
```

```python
df.to_excel('FRANC_ZONE.xlsx')
```

```python
df.to_pickle("FRANC_ZONE.pkl")
```
